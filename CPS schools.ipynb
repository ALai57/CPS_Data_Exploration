```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import datasets to explore
df_2017_12 = pd.read_excel('EmployeePositionRoster_12312017.xls', header=0)
df_2016_12 = pd.read_excel('EmployeePositionRoster_12312016.xls', header=0)
df_2015_12 = pd.read_excel('EmployeePositionRoster_12312015.xls', header=0)
```

```python
# Examine datasets
print(df_2017_12.head())
print(df_2016_12.head())
print(df_2015_12.head())
```

```python
df_2016_12.groupby(['Dept ID', 'Department', 'Job Title'])['Annual Salary', 'Annual Benefit Cost'].agg(['count','mean'])
```

```python
df_2015_12.groupby(['UNIT NUMBER', 'UNIT NAME', 'JOB DESCRIPTION'])['ANNUAL SALARY', 'ANNUAL BENEFIT COST'].agg(['count','mean'])
```

```python
# Look at results by department
df_2017_sum = df_2017_12.groupby(['Dept ID', 'Department'])['Annual Salary', 'Annual Benefit Cost'].agg(['count','sum']).reset_index()
df_2016_sum = df_2016_12.groupby(['Dept ID', 'Department'])['Annual Salary', 'Annual Benefit Cost'].agg(['count','sum']).reset_index()
df_2015_sum = df_2015_12.groupby(['UNIT NUMBER', 'UNIT NAME'])['ANNUAL SALARY', 'ANNUAL BENEFIT COST'].agg(['count','sum']).reset_index()
```

```python
df_2017_sum.columns = ['Dept ID', 'Department', '2017 n_emps','2017 salary','d','2017 benefits']
df_2016_sum.columns = ['Dept ID', 'Department','2016 n_emps','2016 salary','d','2016 benefits']
df_2015_sum.columns = ['Dept ID', 'Department','2015 n_emps','2015 salary','d','2015 benefits']

df_2017_sum.drop('d', axis=1, inplace=True)
df_2016_sum.drop('d', axis=1, inplace=True)
df_2015_sum.drop('d', axis=1, inplace=True)
```

```python
df_full = df_2017_sum.merge(df_2016_sum, on='Dept ID', how='outer').merge(df_2015_sum, on='Dept ID', how='outer')
df_full.drop(['Department_y','Department'], axis=1, inplace= True)

df_full['2017 total'] = df_full['2017 salary'] + df_full['2017 benefits']
df_full['2016 total'] = df_full['2016 salary'] + df_full['2016 benefits']
df_full['2015 total'] = df_full['2015 salary'] + df_full['2015 benefits']

df_full['2017 delta'] = df_full['2017 total'] - df_full['2016 total']
df_full['2016 delta'] = df_full['2016 total'] - df_full['2015 total']
```

```python
print(df_full['2015 total'].sum())
print(df_full['2016 total'].sum())
print(df_full['2017 total'].sum())
```

```python
growth_15_16 = df_full['2016 total'] > df_full['2015 total']
growth_16_17 = df_full['2017 total'] > df_full['2016 total']
```
